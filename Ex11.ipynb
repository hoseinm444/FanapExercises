```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
def show_img_plt(img, fig_height=8, fig_width=6, vmin=0, vmax=255):
    plt.figure(figsize=(fig_height, fig_width), dpi=100)
    plt.imshow(img,cmap='gray', vmin=vmin, vmax=vmax)
```

```python
def show_mult_img_plt(rows, columns, img_names, vmin=0, vmax=255):
    fig = plt.figure(figsize=(15, 17), dpi=100)
    for i in range(len(img_names)):
        fig.add_subplot(rows, columns, i+1)
        plt.imshow(img_names[i],cmap='gray', vmin=vmin, vmax=vmax)
        plt.axis('off')
        plt.title('img_' + str(i))
```

##Ex01

Find and plot the Circles in the image (b.jpg), for pixels with intensity greater than 120.

```python
img = cv2.imread('b.jpg', cv2.IMREAD_COLOR)
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
show_img_plt(gray_img)
imgh,imgw = gray_img.shape
print(imgh,imgw)
```

```python
gauss_img = cv2.GaussianBlur(gray_img, (5, 5), 0)
#canny_edges = cv2.Canny(gauss_img, 120, 200)
circles = cv2.HoughCircles(gauss_img, cv2.HOUGH_GRADIENT,1, img.shape[0]/60,
                           param1=180, param2=170, minRadius=180, maxRadius=0)
```

```python
print(circles)
```

```python
if circles is not None:
    circles = np.uint16(np.around(circles))
    for i in circles[0]:
        cv2.circle(img, (i[0], i[1]), i[2], (0, 255, 0), 2)
show_img_plt(img)
#cv2_imshow(gauss_img)
```

##02
Find the shape of the following object using contours


```python
img_contour = cv2.imread('F6.png', cv2.IMREAD_COLOR)
gray_contour = cv2.cvtColor(img_contour, cv2.COLOR_BGR2GRAY)
#show_img_plt(img_contour)

gauss_contour = cv2.GaussianBlur(gray_contour, (5, 5), 0)
edges_contour = cv2.Canny(gauss_contour, 190, 200)

contours, hierarchy = cv2.findContours(edges_contour, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
cv2.drawContours(img_contour, contours, -1, (255, 0, 0), 3)
```
