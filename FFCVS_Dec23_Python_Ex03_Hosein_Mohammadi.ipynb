##Exercise 15


Implement following architecture using OOP!

use __str__ to print output same as: Breed of the dog is {Breed}, It is {Size} and {Color} , the dog is {Age} years old



```python
class Dog:
    # attribute and constructor
    def __init__(self, Name,Breed,Size,Age,Color):
        self.Dog_name = Name
        self.Dog_breed = Breed
        self.Dog_size = Size
        self.Dog_age = Age
        self.Dog_color = Color
    def Eat():
       print("It is eating.")
    def Sleep():
       print("It is sleeping.")
    def Sit():
       print("It is siting.")
    def Run():
       print("It is running.")
    def __str__(self):
       return f"Breed of the dog({self.Dog_name}) is {self.Dog_breed}, It is {self.Dog_size} and {self.Dog_age} , the dog is {self.Dog_color} years old."

# Object instantiation
Bella = Dog("Bella","Beaglier","Small",2,"red")
print(f'{Bella}')

Max = Dog("Max","Foxhound","big",3,"white")
print(f'{Max}')

Cooper = Dog("Bella","Beagle","Medium",2,"Brown")
print(f'{Cooper}')
```
