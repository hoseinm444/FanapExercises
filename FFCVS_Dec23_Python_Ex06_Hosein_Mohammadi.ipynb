```python
#1
#yes .this picture needs image enhancement
import cv2
import matplotlib.pyplot as plt
import numpy as np
#reading image in OPENCV
img = cv2.imread(r'sample.jpg')

#showing image using matplotlib
def show_img_plt(img, c_map ='gray', fig_hight=8, fig_width=6):
    plt.figure(figsize=(fig_hight, fig_width), dpi=100)
    plt.imshow(img, cmap=c_map)

show_img_plt(img)
```

```python
#display hist using matplotlib
def display_hist_plt(img, bins=256, range=(0, 256)):
    plt.figure(figsize=(4, 2), dpi=100)
    plt.hist(img.flat, bins=bins, range=range)
    plt.show()
display_hist_plt(img)
```

##2

```python

#Contrast Stretching
min_new = 0
max_new = 255

min_old = np.min(img)
max_old = np.max(img)

cs_img = np.uint8(((img-min_old)/(max_old-min_old))*255)
show_img_plt(cs_img)
```

```python
#Histogram Equalization
img = cv2.imread(r'sample.jpg')
grayimg = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
hi_eq = cv2.equalizeHist(grayimg)
display_hist_plt(hi_eq)
```

```python
show_img_plt(hi_eq)
```

```python
#Gamma
gamma = 0.3
gamma_img = np.uint8(((img/255)**gamma)*255)
show_img_plt(gamma_img)
```

##3

 در نهایت پس از اعمال روش گاما آن را با روش هیستوگرام ایکوالیزیشن دوباره اعمال می کنیم تا تصویر بهتری داشته باشیم.

```python
g_g_img = cv2.cvtColor(gamma_img, cv2.COLOR_BGR2GRAY)
hi_eq_gamma = cv2.equalizeHist(g_g_img)
show_img_plt(hi_eq_gamma)
```

```python
display_hist_plt(hi_eq_gamma)
```
